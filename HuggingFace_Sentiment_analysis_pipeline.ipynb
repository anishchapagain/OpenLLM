<h2>Sentiment analysis: Huggingface - transformers</h2>

```python
from transformers import pipeline
```

```python
# Default Model

sentiment_pipeline = pipeline("sentiment-analysis") # model = 'distilbert-base-uncased-finetuned-sst-2-english',
```

## If no model is supplied, defaulted to distilbert/distilbert-base-uncased-finetuned-sst-2-english


```python
# No model was supplied, defaulted to distilbert/distilbert-base-uncased-finetuned-sst-2-english

# Providing the model
sentiment_distillbert = pipeline(task="sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")

sentiment_distillbert("I'm not saying that the output looks okay") # Check the sentiment on provided text
```

```python
reviews = [
    "The staff looks okay but was not quite happy with the service they provided",
    "There's no way that the breakfast was good",
    "I'm not sure for anything, everything was good almost",
    "I'm happy with almost all things",
    "Oh what the things where",
    "I'm not saying that the output looks okay",
    "This is great, wow i'm very happy"
]
```

```python
sentiment_pipeline(reviews) # will show sentiment for all the review available inside reviews.
```

```python
sentiment_distillbert(reviews) # verifying with the model provided
```

# Classification with multiple labels


```python
# https://huggingface.co/SamLowe/roberta-base-go_emotions

multilabel_classifier = pipeline(task="text-classification", model="SamLowe/roberta-base-go_emotions")

# Testing
print(reviews[5])
multilabel_classifier(reviews[5])
```

```python
# Classification with multiple labels

multilabel_classifier = pipeline(task="text-classification", model="SamLowe/roberta-base-go_emotions", top_k=10) # top_k: number of labels to return, None: will return all

# Testing
print(reviews[5]) # {'label': 'NEGATIVE', 'score': 0.9976693987846375} | {'label': 'NEU', 'score': 0.5532468557357788},

multilabel_classifier(reviews[5]) # neutral
```

## Display all labels 'top-k'


```python

multilabel_classifier = pipeline(task="text-classification", model="SamLowe/roberta-base-go_emotions", top_k=None)

print(reviews[2]) #  {'label': 'POSITIVE', 'score': 0.9951435327529907} |  {'label': 'POS', 'score': 0.7109293937683105},

multilabel_classifier(reviews[3]) # joy
```
